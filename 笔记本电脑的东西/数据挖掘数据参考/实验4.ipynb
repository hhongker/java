```python
import numpy as np
# (1)
datas=np.load('D:/root/Desktop/各科/kk/Matplotlib数据及代码/6_Matplotlib_populations.npz',allow_pickle=True)
print(datas)
```

```python
a=datas['data']
print(a)
a = a[0:len(a)-2,:]
```

```python
b=datas['feature_names']
print(b)
```

```python
# (2)
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2,1,1)


#(3)
for i in np.arange(1,6,1):
    ax1.scatter(a[:,0],a[:,i])
plt.title('人口数据')
plt.xlabel(b[0])
c = [i for i in b[1:]]
plt.legend(c,loc='center left')

# (4)
ax2 = fig.add_subplot(2,1,2)
for i in np.arange(1,6,1):
    ax2.plot(a[:,0],a[::-1,i],marker='o')
plt.title('人口数据')
plt.xticks(range(0,20,1),a[::-1,0],rotation=45)
plt.xlabel(b[0])
# c = [i for i in b[1:]]
plt.legend(b[1:],loc='best')
plt.show()
```

```python
# (1)
fig1 = plt.figure(figsize=(14,8))
for i in np.arange(2,4,1):
    plt.bar(a[:,0],a[:,i])
plt.legend(b[2:4])
# 分析：男性人口普遍比女性人口高，而且在增长上幅度比较低，总的来说比较平坦
```

```python
# (2)
fig2 = plt.figure(figsize=(14,8))
for i in np.arange(4,6,1):
    plt.bar(a[:,0],a[:,i],alpha=0.6)
plt.legend(b[4:6])
# 随着时间发展，乡村人口呈现下降趋势，城镇人口呈上升趋势
```
