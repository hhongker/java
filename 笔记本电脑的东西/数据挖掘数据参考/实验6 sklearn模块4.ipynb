### 4. wine_quality数据集的葡萄酒评分在1到10之间，要实现以下步骤：

#### 1）把数据划分为训练集和测试集，并进行标准化，然后分别构建线性回归模型和梯度提升回归模型，y为葡萄酒评分，x为各种化学成分。

```python
import pandas as pd
wine_quality = pd.read_excel("D:/root/Desktop/各科/kk/第6章 使用scikit-learn泰迪代码和数据/sklearn模块4.xlsm",encoding="utf-8")
wine_quality
```

```python
x=wine_quality.iloc[:,:-1]
y=wine_quality['quality']

x,y
```

#### 1）把数据划分为训练集和测试集，并进行标准化，然后分别构建线性回归模型和梯度提升回归模型，y为葡萄酒评分，x为各种化学成分

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=6)
from sklearn.preprocessing import MinMaxScaler #标准化
wine_quality_model = MinMaxScaler().fit(x_train) 
x_trainScaler = wine_quality_model.transform(x_train)
x_testScaler = wine_quality_model.transform(x_test)

from sklearn.linear_model import LinearRegression
clf = LinearRegression().fit(x_trainScaler,y_train)#建立线性回归模型
from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor().fit(x_trainScaler,y_train)#建立梯度回归模型
```

#### 2）分别用两个回归模型对测试样本进行预测评分，并计算和真实评分的均方误差、中值绝对差、可解释方差值。根据得分判断性能优劣。

```python
y_pred1=clf.predict(x_testScaler)
print('线性回归模型预测的结果为：','\n',y_pred1)
y_pred2= GBR.predict(x_testScaler)
print('梯度提升回归模型预测的结果为：','\n',y_pred2)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import explained_variance_score
a=mean_squared_error(y_test,y_pred1)
print('winequality数据线性回归模型的均方误差为：',a)
b=median_absolute_error(y_test,y_pred1)
print('winequality数据线性回归模型的中值绝对误差为：',b)
c=explained_variance_score(y_test,y_pred1)
print('winequality数据线性回归模型的可解释方差值为：',c)

a2=mean_squared_error(y_test,y_pred2)
print('winequality数据梯度提升回归树模型的均方误差为：',a2)
b2=median_absolute_error(y_test,y_pred2)
print('winequality数据梯度提升回归树模型的中值绝对误差为：',b2)
c2=explained_variance_score(y_test,y_pred2)
print('winequality数据梯度提升回归树模型的可解释方差值为：',c2)
```
