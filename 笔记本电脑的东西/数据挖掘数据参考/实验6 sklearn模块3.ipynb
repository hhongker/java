## 二：内容和方法
3. wine数据集的葡萄酒总共分为3种，要实现以下步骤：<br>
（1）按一定比例划分训练集和测试集，进行标准化。<br>
（2）构建SVM分类模型，并预测测试集的分类结果。<br>

（4）把数据进行PCA变换到2维，在平面上用3种不同类型的点绘制出训练样本的分布图。构建构建SVM分类模型，并预测测试集的分类结果,在同一个平面上用另外3种不同类型的点绘制正确分类的测试样本，再用另外3种不同类型的点绘制错误分类的测试样本。

```python
import pandas as pd
datas = pd.read_excel("D:/root/Desktop/各科/kk/7/工作表 在 实验6+sklearn模块3.xlsm",encoding="utf8")
datas
```

#### 1）按一定比例划分训练集和测试集，进行标准化。

```python
# 划分训练集和测试集
y = datas["Class"]
x = datas[datas.columns[1:]]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
print(x_train, x_test, y_train,y_test)
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
model = StandardScaler().fit(x_train)

x_train_ss = model.transform(x_train)
x_test_ss = model.transform(x_test)
print(x_train_ss,x_test_ss)
```

#### 2）构建SVM分类模型，并预测测试集的分类结果。

```python
from sklearn.svm import SVC
model = SVC().fit(x_train_ss,y_train)
y_pre = model.predict(x_test_ss)
y_pre
```

```python
model.score(x_test_ss,y_test)
```

#### 3）打印出分类报告，评价分类模型的性能。

```python
import numpy as np
from sklearn.metrics import recall_score,precision_score,accuracy_score,cohen_kappa_score
from sklearn.metrics import f1_score

print(accuracy_score(y_test,y_pre))
print(recall_score(np.array(y_test),y_pre,average='macro'))
print(precision_score(np.array(y_test),y_pre,average='macro'))
print(f1_score(y_test,y_pre,average='macro'))
print(cohen_kappa_score(y_test,y_pre))
```

#### 4）把数据进行PCA变换到2维，在平面上用3种不同类型的点绘制出训练样本的分布图。构建构建SVM分类模型，并预测测试集的分类结果,在同一个平面上用另外3种不同类型的点绘制正确分类的测试样本，再用另外3种不同类型的点绘制错误分类的测试样本。

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2).fit(x_train)
x_train_pca = model.transform(x_train)
x_test_pca = model.transform(x_test)

x_train_pca,x_test_pca
```

```python
model1=SVC().fit(x_train_pca,y_train)
y_pred_pca=model1.predict(x_test_pca) #预测集
print('y_test:',list(y_test))
print('y_pre:',list(y_pred_pca))
```
