## 1.对给定数据集5_pandas_data.csv，完成如下功能：
1) 判断第一列(Id)是否有缺失值，如果有，则补全。<br>
2) 判断是否有重复记录，如果有，则删除至唯一。<br>
3) 计算成绩的平均值，作为新的一列加入到原数据中。<br>
4) 寻找平均分最高的记录。<br>
5) 统计每个科目90分以上的人数。

```python
import numpy as np
import pandas as pd
```

```python
# 1) 判断第一列(Id)是否有缺失值，如果有，则补全。
a = pd.read_csv('D:/root/Desktop/各科/kk/5_pandas_data.csv.csv',encoding='utf8')
print(a)
loc_list = list(a[a['Id'].isnull()].index)
for i in loc_list:
    a['Id'][i] = i + 1
# a = a.fillna(13)
print(a)
```

```python
# 2) 判断是否有重复记录，如果有，则删除至唯一。
a = a.drop_duplicates(['Id'])
a
```

```python
# 3) 计算成绩的平均值，作为新的一列加入到原数据中。
a['mean'] = a[['Chinese','Math','English']].agg('mean',axis=1)
a
```

```python
# 4) 寻找平均分最高的记录。
np.max(a['mean'])
```

```python
#（5）统计每个科目90分以上的人数。
a[(a[['Chinese','Math','English']]>90).all(axis=1)].index.size
```

```python
# 6) 被下单最多商品(item)是什么？
# data_gb=data.groupby(['item_name']).agg({'quantity':'sum'})
# data_gb[:][data_gb['quantity']==np.max(data_gb['quantity'])]
data_gb = data[['item_name','quantity']].groupby(['item_name'])
data_gb_after=data_gb.sum().reset_index()
data_gb_after[:][data_gb_after['quantity']==np.max(data_gb_after['quantity'])]
```

```python
# 9) 将item_price转换为浮点数。
data['item_price'] = data['item_price'].agg(lambda x:float(x.replace('$','')))
data['item_price']
```

```python
# 10) 在该数据集对应的时期内，收入(revenue)是多少？
np.sum(data['item_price']*data['quantity'])
```

```python
# 11) 在改数据集对应的时期内，一共有多少订单？
len(data.groupby('order_id'))
```

```python
# 12) 每一单(order)对应的平均总价是多少？
pd.pivot_table(data[['order_id','item_price']],
      index ='order_id',aggfunc = ['mean','sum'],
      margins = True)
```
